```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import TruncatedSVD
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
df_train = pd.read_csv('data/train_v2.csv')
df_test = pd.read_csv('data/test_v2.csv')
y_train = df_train['rating']
```

###Put all into a matrix

```python
matrix = pd.concat([df_train, df_test]).pivot('user', 'movie', 'rating')
matrix.head()
```

```python
movie_means = matrix.mean()
user_means = matrix.mean(axis=1)
mzm = matrix - movie_means #zero-mean method
mz = mzm.fillna(0) #fill all to become zero
mask = -mzm.isnull() #mask those that is null
mz.head()
```

###Repeated Matrix Reconstruction

```python
%%time

iteration = 0 #initialize
mse_last = 999 #initialize
while iteration < 10:
    iteration += 1
    
    #Using SVD to do matrix transformation
    svd = TruncatedSVD(n_components=15, random_state=42)
    svd.fit(mz)
    
    #transform back to original space and put into dataframe
    mzsvd = pd.DataFrame(svd.inverse_transform(svd.transform(mz)),
                      columns=mz.columns, index=mz.index)
    
    #calculate MSE
    mz_pred = mzsvd[mask].fillna(0)
    mz_true = mzm[mask].fillna(0)
    mse = mean_squared_error(mz_true, mz_pred) 
    print('%i, %.5f, %.5f' % (iteration, mse, mse_last-mse))
    
    #update to latest matrix
    mzsvd[mask] = mzm[mask]
    mz = mzsvd
    
    #break if mse is less than 0.00001
    if mse_last - mse < 0.00001: break
    mse_last = mse
```

```python
m = mz + movie_means
m = m.clip(lower=1, upper=5)
m.head(10)
```

```python
%%time 

#search for rating from the matrix generated
df_test['rating'] = df_test.apply(lambda x:m[m.index==x.user][x.movie].values[0],axis=1)
df_test.head()
```

```python
missing = np.where(df_test.rating.isnull())[0]
df_test.ix[missing,'rating'] = user_means[df_test.loc[missing].user].values

df_test.to_csv('submission.csv', index=False, columns=['ID','rating'])
```
