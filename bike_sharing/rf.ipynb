###Random Forests

```python
def rmsle(actual_values, predicted_values):
    assert len(actual_values) == len(predicted_values)
    actual_values = np.ravel(actual_values)
    predicted_values = np.ravel(predicted_values)
    total = 0
    for a, p in zip(actual_values, predicted_values):
        total += math.pow(math.log(p+1) - math.log(a+1), 2)
    return math.sqrt(total/len(actual_values))
```

```python
%%time 

clf = RandomForestClassifier(n_estimators = 500, n_jobs=4).fit(X_train, y_train)

predicted = clf.predict(X_test)
print "RMSLE:", rmsle(y_test, predicted)
```

```python
%%time

params = {'n_estimators': [100, 200, 500, 1000],
#           "max_depth": [3, None],
#           "max_features": [1, 3, 10],
#           "min_samples_split": [1, 3, 10],
#           "min_samples_leaf": [1, 3, 10],
#           "bootstrap": [True, False],
#           "criterion": ["gini", "entropy"]
          }

est = RandomForestClassifier(n_estimators=500, n_jobs=4)

n_subsamples = 500
X_small_train, y_small_train = X_train[:n_subsamples], y_train[:n_subsamples]

gs = GridSearchCV(est, params, cv=3,
                  n_jobs=4,
                  verbose=1)

_ = gs.fit(X_small_train, y_small_train)

print "GS Best Score: %0.2f" % gs.best_score_
print "GS Best Params:", gs.best_params_
```

###Accuracy Score

```python
est.set_params(**gs.best_params_)
est.fit(X_train, y_train)

predicted = est.predict(X_test)
print "RMSLE:", rmsle(y_test, predicted)
```
